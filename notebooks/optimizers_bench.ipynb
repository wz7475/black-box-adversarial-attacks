```python
import matplotlib.pyplot as plt
import numpy as np

# Check if alpha column exists, if not we need to extract it from another column
print(df.columns.tolist())
df.head()
```

```python
fig, axes = plt.subplots(3, 1, figsize=(14, 12))

eps_values = [0.01, 0.1, 0.2]
alpha_values = sorted(df['alpha'].unique())
colors = plt.cm.Set2(np.linspace(0, 1, len(alpha_values)))

for idx, eps in enumerate(eps_values):
    ax = axes[idx]
    
    # Filter data for current eps
    df_eps = df[df['eps'] == eps]
    
    # Get unique optimizers and sort by mean success rate
    optimizers = df_eps['optimizer_name'].unique()
    
    # Calculate mean success rate for each optimizer across all alphas
    mean_success_by_opt = {}
    for opt in optimizers:
        mean_success = df_eps[df_eps['optimizer_name'] == opt]['success_ratio'].mean()
        mean_success_by_opt[opt] = mean_success
    
    # Sort optimizers by mean success rate in descending order
    optimizers = sorted(optimizers, key=lambda x: mean_success_by_opt[x], reverse=True)
    
    # Set up bar positions
    x = np.arange(len(optimizers))
    width = 0.25  # width of each bar
    
    # Plot bars for each alpha
    for i, alpha in enumerate(alpha_values):
        df_alpha = df_eps[df_eps['alpha'] == alpha]
        
        # Create a list of success ratios in the same order as optimizers
        success_ratios = []
        for opt in optimizers:
            opt_data = df_alpha[df_alpha['optimizer_name'] == opt]
            if len(opt_data) > 0:
                success_ratios.append(opt_data['success_ratio'].values[0])
            else:
                success_ratios.append(0)
        
        # Plot bars
        offset = (i - len(alpha_values)/2 + 0.5) * width
        ax.bar(x + offset, success_ratios, width, label=f'={alpha}', color=colors[i])
    
    # Customize subplot
    ax.set_xlabel('Optimizer', fontsize=12)
    ax.set_ylabel('Success Rate', fontsize=12)
    ax.set_title(f'Success Rates for ={eps}', fontsize=14, fontweight='bold')
    ax.set_xticks(x)
    ax.set_xticklabels(optimizers, rotation=45, ha='right')
    ax.legend()
    ax.grid(axis='y', alpha=0.3)
    # ax.set_ylim([0, max(df['success_ratio'].max() * 1.1, 0.1)])

plt.tight_layout()
plt.show()
```
